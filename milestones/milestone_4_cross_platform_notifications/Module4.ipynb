## Scrape Book Information

### Subtask:
Scrape the UPC codes, book titles, and prices from all pages of 'https://books.toscrape.com/'.


**Reasoning**:
The first step in scraping is to import the necessary libraries, `requests` for handling HTTP requests and `BeautifulSoup` for parsing HTML, then define the base URL and an empty list to store the scraped data.

````python
import requests
from bs4 import BeautifulSoup

# Initialize an empty list to store book data
book_data = []

# Define the base URL and the initial URL
base_url = 'https://books.toscrape.com/'
current_url = 'https://books.toscrape.com/'

print("Libraries imported, book_data list initialized, and URLs defined.")
````

**Reasoning**:
Now that the libraries are imported and variables are initialized, the next step is to implement the scraping logic, which involves iterating through all pages, extracting book details (title, price, UPC), and storing them in the `book_data` list.



````python
while current_url:
    print(f"Scraping page: {current_url}")
    response = requests.get(current_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all articles for books on the current page
    articles = soup.find_all('article', class_='product_pod')

    for article in articles:
        # Extract title and price
        title = article.h3.a['title']
        price = article.find('p', class_='price_color').text.strip()

        # Get the relative link to the book's detail page
        relative_link = article.h3.a['href']
        if 'catalogue/' not in relative_link:
            book_detail_url = base_url + 'catalogue/' + relative_link
        else:
            book_detail_url = base_url + relative_link

        # Go to the book's detail page to get the UPC
        detail_response = requests.get(book_detail_url)
        detail_soup = BeautifulSoup(detail_response.text, 'html.parser')

        # Find the UPC code (usually in a table with product information)
        upc_element = detail_soup.find('th', string='UPC')
        if upc_element:
            upc = upc_element.find_next_sibling('td').text
        else:
            upc = 'N/A' # In case UPC is not found

        book_data.append({
            'UPC': upc,
            'Title': title,
            'Price': price
        })

    # Find the 'next' button to go to the next page
    next_button = soup.find('li', class_='next')
    if next_button:
        next_page_relative_url = next_button.a['href']
        if 'catalogue/' not in next_page_relative_url:
            current_url = base_url + 'catalogue/' + next_page_relative_url
        else:
            current_url = base_url + next_page_relative_url
    else:
        current_url = None # No more pages, stop the loop

print("Scraping complete. Total books scraped:", len(book_data))
````

````python
import requests
import json

# 2. Define the API key
API_KEY = '6nszyk9boeq6yad0hqh4'

# 3. Initialize an empty list to store API call results
api_results = []

# Base URL for the booksrun.com API
BOOKS_RUN_API_BASE_URL = 'https://booksrun.com/api/v3/price/buy/'

print("Starting API calls for mock ISBNs...")

# 4. Loop through each isbn in the mock_isbns list
for isbn in mock_isbns:
    # 5. Construct the full API URL
    api_url = f"{BOOKS_RUN_API_BASE_URL}{isbn}?key={API_KEY}"

    raw_json_response = {}
    extracted_price = 'Price not found or error occurred'

    try:
        # 6. Make an HTTP GET request
        response = requests.get(api_url)
        response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

        # 7. Parse the JSON response received from the API
        raw_json_response = response.json()

        # 8. Extract the competitor's price
        # The structure can vary, so we need to check existence carefully
        # Assuming price is under 'result' -> 'offers' -> first item -> 'new_price' or 'used_price'
        if 'result' in raw_json_response and 'offers' in raw_json_response['result'] and raw_json_response['result']['offers']:
            # Try to get new_price first, then used_price
            offer = raw_json_response['result']['offers'][0]
            if 'new_price' in offer and offer['new_price'] is not None:
                extracted_price = f"New: ${offer['new_price']}"
            elif 'used_price' in offer and offer['used_price'] is not None:
                extracted_price = f"Used: ${offer['used_price']}"
            else:
                extracted_price = 'No specific price found in offer'
        else:
            extracted_price = 'No offers available'

    except requests.exceptions.RequestException as e:
        extracted_price = f"Request failed: {e}"
        print(f"Error fetching data for ISBN {isbn}: {e}")
    except json.JSONDecodeError:
        extracted_price = "Invalid JSON response"
        print(f"Error decoding JSON for ISBN {isbn}: Invalid JSON response")
    except Exception as e:
        extracted_price = f"An unexpected error occurred: {e}"
        print(f"Unexpected error for ISBN {isbn}: {e}")

    # 9. Store the mock ISBN, the full API URL, the raw JSON response, and the extracted price
    api_results.append({
        'Mock ISBN': isbn,
        'API URL': api_url,
        'Raw JSON Response': raw_json_response,
        'Extracted Price': extracted_price
    })

print("API calls complete. Results stored in api_results list.")
````

````python
import requests
import json

# API Key for booksrun.com
API_KEY = '6nszyk9boeq6yad0hqh4'
BOOKS_RUN_API_BASE_URL = 'https://booksrun.com/api/v3/price/buy/'

# List to store results from booksrun.com API calls using real ISBNs
api_results_real_isbns = []

print("Making API calls to booksrun.com with real ISBNs (with corrected price extraction)...")

for isbn_to_query in real_isbns:
    api_url_used = f"{BOOKS_RUN_API_BASE_URL}{isbn_to_query}?key={API_KEY}"

    raw_json_response_real = {}
    extracted_price_real = 'Price not found or error occurred'

    try:
        response_real = requests.get(api_url_used)
        response_real.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

        raw_json_response_real = response_real.json()

        # Extract the competitor's price with improved logic
        if 'result' in raw_json_response_real and raw_json_response_real['result'].get('status') == 'success':
            offers_data = raw_json_response_real['result'].get('offers', {})

            # Prioritize booksrun offers
            booksrun_offers = offers_data.get('booksrun', {})
            if booksrun_offers and booksrun_offers.get('new') and booksrun_offers['new'] != 'none':
                extracted_price_real = f"New: ${booksrun_offers['new']['price']}"
            elif booksrun_offers and booksrun_offers.get('used') and booksrun_offers['used'] != 'none':
                extracted_price_real = f"Used: ${booksrun_offers['used']['price']}"
            else:
                # Check marketplace offers if booksrun offers are not found or 'none'
                marketplace_offers = offers_data.get('marketplace', [])
                if isinstance(marketplace_offers, list) and len(marketplace_offers) > 0:
                    first_marketplace_offer = marketplace_offers[0]
                    if first_marketplace_offer.get('new') and first_marketplace_offer['new'] != 'none':
                        extracted_price_real = f"New Marketplace: ${first_marketplace_offer['new']['price']}"
                    elif first_marketplace_offer.get('used') and first_marketplace_offer['used'] != 'none':
                        extracted_price_real = f"Used Marketplace: ${first_marketplace_offer['used']['price']}"
                    else:
                        extracted_price_real = 'No specific price found in marketplace offer'
                else:
                    extracted_price_real = 'No offers available'
        else:
            extracted_price_real = raw_json_response_real['result'].get('message', 'API call failed')

    except requests.exceptions.RequestException as e:
        extracted_price_real = f"Request failed: {e}"
        print(f"Error fetching data for ISBN {isbn_to_query}: {e}")
    except json.JSONDecodeError:
        extracted_price_real = "Invalid JSON response"
        print(f"Error decoding JSON for ISBN {isbn_to_query}: Invalid JSON response")
    except Exception as e:
        extracted_price_real = f"An unexpected error occurred: {type(e).__name__} - {e}"
        print(f"Unexpected error for ISBN {isbn_to_query}: {e}")

    api_results_real_isbns.append({
        'ISBN Used': isbn_to_query,
        'API URL': api_url_used,
        'Raw JSON Response': raw_json_response_real,
        'Extracted Price': extracted_price_real
    })

print("API calls to booksrun.com complete for real ISBNs.")

# Display the results
print("\nDisplaying API call results for real ISBNs:")
for result in api_results_real_isbns:
    print("\n-----------------------------------------")
    print(f"ISBN Used: {result['ISBN Used']}")
    print(f"API URL: {result['API URL']}")
    print(f"Raw JSON Response: {json.dumps(result['Raw JSON Response'], indent=2)}")
    print(f"Extracted Price: {result['Extracted Price']}")
print("\n-----------------------------------------")
````

````python
import numpy as np

def suggest_price_action_dynamic(row, competitive_margin_below=0.95, competitive_margin_above=1.05):
    """
    Suggests a dynamic price action based on competitor prices, aiming for real-time profit optimization.

    This function provides a more nuanced approach than simply 'reduce' or 'increase'.
    It attempts to suggest a target price that either makes the product competitive
    or maximizes profit when the original price is already lower than the competitor's.

    Args:
        row (pd.Series): A row from the DataFrame containing 'Original Price (Float)' and 'Competitor Price (Float)'.
        competitive_margin_below (float): Factor to apply to competitor's price when original is lower.
                                          E.g., 0.95 suggests setting price at 95% of competitor's price for profit boost.
        competitive_margin_above (float): Factor to apply to competitor's price when original is higher.
                                          E.g., 1.05 suggests setting price at 105% of competitor's price
                                          to stay competitive while preserving some margin.

    Returns:
        str: A string describing the suggested price action and a potential target price.
    """
    original_price = row['Original Price (Float)']
    competitor_price = row['Competitor Price (Float)']

    # If prices are not available from either source, no action can be taken.
    if pd.isna(original_price) or pd.isna(competitor_price):
        return 'No Action (Price not available from one/both sources)'

    # Define a small tolerance for price equality to avoid unnecessary micro-adjustments.
    # Using a relative tolerance (e.g., 1% of the original price) is often more robust.
    price_equality_tolerance = 0.01 * original_price # 1% of original price

    if original_price > competitor_price + price_equality_tolerance:
        # Scenario 1: Original price is significantly higher than competitor's price.
        # Strategy: Reduce price to be competitive. A target could be slightly above the competitor's price
        # to maintain some margin, or directly at the competitor's price.
        # We calculate a suggested new price that is 'competitive_margin_above' times the competitor's price.
        # However, we ensure this suggested price is not higher than the current original price.
        suggested_new_price = min(original_price, competitor_price * competitive_margin_above)
        return f'Reduce Price for Competitiveness: Target approx. ${suggested_new_price:.2f} (from ${original_price:.2f})'
    elif original_price < competitor_price - price_equality_tolerance:
        # Scenario 2: Original price is significantly lower than competitor's price.
        # Strategy: Increase price for profit optimization. A target could be slightly below
        # the competitor's price to maintain a competitive edge, or at the competitor's price.
        # We calculate a suggested new price that is 'competitive_margin_below' times the competitor's price.
        # We ensure this suggested price is not lower than the current original price.
        suggested_new_price = max(original_price, competitor_price * competitive_margin_below)
        return f'Increase Price for Profit Optimization: Target approx. ${suggested_new_price:.2f} (from ${original_price:.2f})'
    else:
        # Scenario 3: Original price is approximately equal to the competitor's price within the tolerance.
        # Strategy: Maintain the current price as it is already competitive and optimal.
        return 'Maintain Price (Competitive & Optimal)'

# Apply the dynamic price adjustment function to create a new column
# This new column provides actionable insights for real-time price changes.
final_comparison_df['Suggested Price Action (Dynamic)'] = final_comparison_df.apply(
    lambda row: suggest_price_action_dynamic(row),
    axis=1
)

# Display the first few rows of the updated DataFrame with price differences and the new dynamic suggestions
print("\nFinal Comparison DataFrame with Price Differences and Dynamic Suggested Actions:")
print(final_comparison_df.head())
print(f"Shape of final_comparison_df: {final_comparison_df.shape}")

# Prepare the DataFrame for a clear tabular display of the analysis results.
# This selection includes the book title, original and competitor prices, calculated differences,
# and the new dynamic suggested price action for easy review and decision-making.
final_analysis_df = final_comparison_df[[
    'Book Title',
    'Original Price (Float)',
    'Competitor Price (Float)',
    'Absolute Price Difference',
    'Percentage Price Difference',
    'Suggested Price Action (Dynamic)' # Using the new dynamic action column
]]

print("\nPrice Adjustment Analysis (Dynamic Strategy):")
# Convert the DataFrame to markdown format for a clean, readable table output.
print(final_analysis_df.to_markdown(index=False))
````

Let's break down how the dynamic pricing strategy works when your original price is significantly higher than the competitor's, using **'Sharp Objects'** as an example:

*   **Competitor Price (Float):** $4.99 (BooksRun price)

**Steps the `suggest_price_action_dynamic` function takes:**

1.  **Calculate `price_equality_tolerance`:**
    *   This tolerance helps decide if prices are 'close enough' to be considered equal, avoiding unnecessary tiny adjustments.
    *   It's set as `0.01 * original_price` (1% of your book's current price).
    *   For 'Sharp Objects': `0.01 * $47.82 = $0.4782`

2.  **Check Condition for Price Reduction:**
    *   The function looks for cases where `original_price > competitor_price + price_equality_tolerance`.
    *   For 'Sharp Objects': `$47.82 > $4.99 + $0.4782`
    *   This simplifies to: `$47.82 > $5.4682`. Since this is `True`, the function identifies that your price is significantly higher than the competitor's.

3.  **Determine `suggested_new_price` for Competitiveness:**
    *   **Goal:** Bring your price down to be competitive, but try to retain some profit margin if possible.
    *   The function calculates a potential new price based on the `competitive_margin_above` factor (which defaults to 1.05 or 105%). This means aiming for a price that is 5% *above* the competitor's price.
    *   Calculated target: `competitor_price * 1.05 = $4.99 * 1.05 = $5.2395`
    *   **Crucially:** The function then takes the *minimum* of your current `original_price` and this calculated target. This ensures you never suggest a new price that's *higher* than what you're already charging, especially when your current price is very high.
    *   `suggested_new_price = min($47.82, $5.2395)`
    *   Therefore, `suggested_new_price = $5.2395` (which would typically be rounded to $5.24).

4.  **Formulate the Action:**
    *   Based on these calculations, the function outputs:
        "**Reduce Price for Competitiveness: Target approx. $5.24 (from $47.82)**"

**In Summary:** Even though the tolerance uses a constant factor (0.01), the *result* (`$0.4782` in this case) scales dynamically with the book's price. The overall strategy is dynamic because it looks at the relationship between your price and the competitor's, and then calculates a *specific new target price* that aims to make you competitive (e.g., slightly above the competitor) without undercutting yourself unnecessarily.

## Integrating LLM Insights into Dynamic Price Adjustment

To move beyond static `competitive_margin_above` and `competitive_margin_below` constants, we can leverage LLM insights. An LLM, trained on vast amounts of data including market trends, competitor strategies, product reviews, and economic indicators, could provide more intelligent, context-aware values for these margins.

Instead of:
```python
competitive_margin_above = 1.05
competitive_margin_below = 0.95
```

The LLM could analyze a specific book and determine:
*   "For this high-demand, strong-brand book, we can set `competitive_margin_above = 1.10` (less aggressive reduction needed, higher profit retention) and `competitive_margin_below = 0.98` (a safer, smaller increase, still competitive)."
*   "For this niche market, low-stock book, we need to aggressively match competitor prices if higher, so `competitive_margin_above = 1.00`. If we're cheaper, we can take more profit, so `competitive_margin_below = 1.03`."

This demonstration will simulate an LLM providing these dynamic margins for a few example books. Our `suggest_price_action_dynamic` function will then be updated to consult these 'LLM insights' before falling back to default values.

````python
# Simulate LLM-generated dynamic margins for specific books
# In a real-world scenario, this would be an actual LLM API call
# that takes book details and returns optimized margins.

llm_derived_margins = {
    "Sharp Objects": {"competitive_margin_above": 1.08, "competitive_margin_below": 0.92}, # LLM suggests less aggressive reduction, more aggressive profit-taking
    "The Coming Woman: A Novel Based on the Life of the Infamous Feminist, Victoria Woodhull": {"competitive_margin_above": 1.02, "competitive_margin_below": 0.98}, # LLM suggests tighter margins for this book
    "How Music Works": {"competitive_margin_above": 1.00, "competitive_margin_below": 1.00}, # LLM suggests direct matching for competitiveness/profit
    "The Five Love Languages: How to Express Heartfelt Commitment to Your Mate": {"competitive_margin_above": 1.05, "competitive_margin_below": 1.00}
}

print("Simulated LLM-derived margins created for demonstration.")
print("These margins will now be used by the dynamic pricing function.")
````

````python
import numpy as np

def suggest_price_action_dynamic(row):
    """
    Suggests a dynamic price action based on competitor prices, aiming for real-time profit optimization.

    This function provides a more nuanced approach by adjusting the competitive margins
    based on the magnitude of the price difference between the original and competitor prices.
    It now also integrates LLM-derived margins if available for a given book.

    Args:
        row (pd.Series): A row from the DataFrame containing 'Book Title', 'Original Price (Float)' and 'Competitor Price (Float)'.

    Returns:
        str: A string describing the suggested price action and a potential target price.
    """
    original_price = row['Original Price (Float)']
    competitor_price = row['Competitor Price (Float)']
    book_title = row['Book Title']

    # Fetch LLM-derived margins if available for this specific book
    # Default to previously defined tiered logic if LLM doesn't have specific insights for this book
    llm_margins = llm_derived_margins.get(book_title, {})

    # Define base margins (can be fixed or default tiered)
    dynamic_margin_above_base = 1.05
    dynamic_margin_below_base = 0.95

    # If prices are not available from either source, no action can be taken.
    if pd.isna(original_price) or pd.isna(competitor_price):
        return 'No Action (Price not available from one/both sources)'

    # Define a small relative tolerance for price equality to avoid unnecessary micro-adjustments.
    price_equality_tolerance = 0.01 * original_price # 1% of original price

    # Calculate percentage difference if original_price is not zero
    percentage_diff = np.nan
    if original_price != 0:
        percentage_diff = ((original_price - competitor_price) / original_price) * 100

    if original_price > competitor_price + price_equality_tolerance:
        # Scenario 1: Original price is significantly higher than competitor's price.
        # Determine margin for reduction dynamically based on how much higher it is.
        dynamic_margin_above = llm_margins.get("competitive_margin_above", dynamic_margin_above_base)
        if not llm_margins: # If no LLM insight, use tiered logic
            if pd.notna(percentage_diff):
                if percentage_diff > 50: # Much higher (e.g., more than 50% pricier)
                    dynamic_margin_above = 1.00 # Aim to match competitor price (aggressive reduction)
                elif percentage_diff > 20: # Moderately higher (e.g., 20-50% pricier)
                    dynamic_margin_above = 1.03 # Aim slightly above competitor
            # else, use default dynamic_margin_above_base

        suggested_new_price = min(original_price, competitor_price * dynamic_margin_above)
        return f'Reduce Price for Competitiveness (LLM/Tiered): Target approx. ${suggested_new_price:.2f} (from ${original_price:.2f})'

    elif original_price < competitor_price - price_equality_tolerance:
        # Scenario 2: Original price is significantly lower than competitor's price.
        # Determine margin for increase dynamically based on how much lower it is.
        dynamic_margin_below = llm_margins.get("competitive_margin_below", dynamic_margin_below_base)
        if not llm_margins: # If no LLM insight, use tiered logic
            if pd.notna(percentage_diff):
                if percentage_diff < -50: # Much lower (e.g., more than 50% cheaper)
                    dynamic_margin_below = 1.00 # Aim to match competitor price (aggressive profit-taking)
                elif percentage_diff < -20: # Moderately lower (e.g., 20-50% cheaper)
                    dynamic_margin_below = 0.97 # Aim slightly below competitor
            # else, use default dynamic_margin_below_base

        suggested_new_price = max(original_price, competitor_price * dynamic_margin_below)
        return f'Increase Price for Profit Optimization (LLM/Tiered): Target approx. ${suggested_new_price:.2f} (from ${original_price:.2f})'
    else:
        # Scenario 3: Original price is approximately equal to the competitor's price within the tolerance.
        # Strategy: Maintain the current price as it is already competitive and optimal.
        return 'Maintain Price (Competitive & Optimal)'

# Apply the dynamic price adjustment function to create a new column
# This new column provides actionable insights for real-time price changes.
final_comparison_df['Suggested Price Action (Dynamic)'] = final_comparison_df.apply(
    lambda row: suggest_price_action_dynamic(row),
    axis=1
)

# Display the first few rows of the updated DataFrame with price differences and the new dynamic suggestions
print("\nFinal Comparison DataFrame with Price Differences and Dynamic Suggested Actions:")
print(final_comparison_df.head())
print(f"Shape of final_comparison_df: {final_comparison_df.shape}")

# Prepare the DataFrame for a clear tabular display of the analysis results.
# This selection includes the book title, original and competitor prices, calculated differences,
# and the new dynamic suggested price action for easy review and decision-making.
final_analysis_df = final_comparison_df[[
    'Book Title',
    'ISBN',
    'Booksrun Link',
    'Original Price (Float)',
    'Competitor Price (Float)',
    'Absolute Price Difference',
    'Percentage Price Difference',
    'Suggested Price Action (Dynamic)' # Using the new dynamic action column
]]

print("\nPrice Adjustment Analysis (Dynamic Strategy):")
# Convert the DataFrame to markdown format for a clean, readable table output.
print(final_analysis_df.to_markdown(index=False))
````

### 3. Task 1: Conversion of UPC Code to ISBN

**Difference between UPC and ISBN:**
*   **UPC (Universal Product Code):** Primarily used in retail to identify consumer products. It's a 12-digit number typically found on packaging and scanned at points of sale. A single book title can have different UPCs depending on the edition, publisher, or even packaging variations.
*   **ISBN (International Standard Book Number):** A unique numeric commercial book identifier. It's a 10-digit (until 2007) or 13-digit (since 2007) number specifically for books and related products. Each edition and format of a book (e.g., paperback, hardcover, e-book) is assigned a different ISBN.

**Why UPC codes cannot directly be used for book pricing APIs:** Book pricing APIs, especially those used by booksellers and aggregators, rely almost exclusively on ISBNs for accurate product identification. UPCs are too generic for books and do not uniquely identify specific editions, which is crucial for precise price comparison. Therefore, a conversion step is essential.

**Method used to extract real ISBNs using Google Books API:**
To bridge the gap between the scraped UPC-identified books and ISBN-based pricing APIs, the Google Books API was employed. The methodology involved using the extracted book titles as search queries to the Google Books API, which then returned relevant book information, including available ISBNs.

**API Request Structure:**
For each book title, an HTTP GET request was constructed to the Google Books API endpoint:
`https://www.googleapis.com/books/v1/volumes`

**Input:** The book's `Title` from the scraped `df_books` DataFrame.
**Parameters:**
*   `q`: The book title (e.g., `'Sharp Objects'`).
*   `maxResults`: Limited to 1 to retrieve the most relevant result.
*   `langRestrict`: Set to `'en'` to prioritize English language results.

**Example API Request (conceptual):**
`requests.get('https://www.googleapis.com/books/v1/volumes', params={'q': 'Sharp Objects', 'maxResults': 1, 'langRestrict': 'en'})`

**Response Parsing Logic:**
Upon receiving a JSON response from the Google Books API, the system parsed the `items` array to find the `volumeInfo` for the top result. Within `volumeInfo`, it looked for `industryIdentifiers`. These identifiers contain a list of dictionaries, each specifying an identifier `type` (e.g., 'ISBN_13', 'ISBN_10') and its `identifier` value. The parsing logic prioritized `ISBN_13` and fell back to `ISBN_10` if `ISBN_13` was not available or valid.

**Error Handling for Missing ISBNs:** If the Google Books API response did not contain any items or valid `industryIdentifiers`, the `ISBN_13` and `ISBN_10` fields were recorded as 'N/A' or 'Error'. This ensured that books without discoverable ISBNs were gracefully handled without halting the process.

**Output generated (list of valid ISBN-10 / ISBN-13):**
The output of this task was a list of dictionaries (`books_with_isbns`), where each dictionary contained the 'Original Title', the extracted 'ISBN_13', 'ISBN_10', and the full 'Google Books API Response' for debugging. From this list, a `real_isbns` list was then compiled containing only the valid and consolidated ISBNs for subsequent competitor API calls.

### 4. Task 2: Web / API Query for Competitor Pricing

**Purpose of querying competitor pricing:** The primary purpose is to gather competitive intelligence that informs dynamic pricing decisions. By comparing our prices against those of a competitor like BooksRun, we can identify opportunities to:
*   Reduce prices to be more competitive and attract more sales.
*   Increase prices for products where we are significantly cheaper, thereby optimizing profit margins without losing competitiveness.
*   Maintain prices if they are already competitive and optimal.

**BooksRun API endpoint structure:**
The BooksRun API provides a specific endpoint for retrieving pricing information for books identified by their ISBN. The structure of the endpoint used is:
`https://booksrun.com/api/v3/price/buy/<ISBN>?key=<API_KEY>`

**Authentication via API key:** Access to the BooksRun API is authenticated using a unique API key, provided as a query parameter (`key=<API_KEY>`). This key ensures authorized access to the service and tracks API usage.

**HTTP GET request logic:**
The system iterates through the `real_isbns` list obtained from Task 1. For each ISBN, a complete API URL is constructed by embedding the ISBN and the `API_KEY`. An HTTP GET request is then sent to this URL using the `requests` library. Error handling (`response.raise_for_status()`) is included to catch HTTP errors (e.g., 4xx, 5xx responses).

**JSON response structure:**
The BooksRun API returns pricing information in a JSON format. A typical successful response includes a `result` object, which contains `status` (e.g., 'success', 'error'), `message`, and an `offers` object. The `offers` object typically differentiates between 'booksrun' (direct offers from the platform) and 'marketplace' (offers from third-party sellers), and within these, 'new' and 'used' prices.

**Extraction of:**
*   **New price:** Prioritized extraction from `offers['booksrun']['new']['price']`.
*   **Used price:** If new price is not available, extracted from `offers['booksrun']['used']['price']`.
*   **Marketplace price:** If BooksRun's direct offers are not available, the system attempts to extract from the first available `offers['marketplace'][0]['new']['price']` or `offers['marketplace'][0]['used']['price']`.

**Handling edge cases:**
*   **Invalid ISBN:** If an ISBN is not recognized by BooksRun, the API returns a status of 'error' and a message like 'Wrong Isbn'. The system captures this message, and the `Extracted Price` is recorded as 'Wrong Isbn' or 'No offers available'.
*   **No offers available:** For valid ISBNs, there might be no current offers (new or used) from BooksRun or its marketplace. In such cases, the `offers` object might contain 'none' values or be empty. The logic gracefully handles these scenarios, reporting 'No offers available'.
*   **API response errors:** Standard `requests.exceptions.RequestException` catches network issues or bad HTTP responses. `json.JSONDecodeError` handles cases where the response is not valid JSON. A general `Exception` catches any other unforeseen errors during the API interaction.

**Input:** `real_isbns` (list of valid ISBNs).
**Output:** A list of dictionaries (`api_results_real_isbns`), each containing:

### 5. Task 3: Price Comparison Between Competitor and Our Website

**Data cleaning:**
Before numerical comparison, the price data, which originally contained currency symbols and descriptive text, required thorough cleaning and normalization. A dedicated function, `clean_and_convert_price`, was developed for this purpose.

**Methodology:**
*   **Removal of Currency Symbols:** Characters such as 'Â£' (for British Pound) and '$' (for US Dollar) were removed from both the 'Original Price (Books.toscrape.com)' and 'Competitor Price (Booksrun.com)' columns. The specific encoding `\u00c2\u00a3` for '£' was handled.

*   **Conversion to Float:** The cleaned price strings were then converted to a floating-point numerical data type. This conversion is essential for performing mathematical operations (subtraction, division) required for price difference calculations.
*   **Handling Non-Numerical Values:** Robust error handling was implemented within the `clean_and_convert_price` function.
