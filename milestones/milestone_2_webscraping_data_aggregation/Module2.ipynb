````python
# Install necessary packages

!pip install playwright nest_asyncio
!playwright install chromium
!apt-get install libatk1.0-0 libatk-bridge2.0-0 libatspi2.0-0 libxcomposite1

# Common imports and Colab event loop fix
import asyncio, json, csv, time
from pathlib import Path
import nest_asyncio
nest_asyncio.apply()
from playwright.async_api import async_playwright

#AJAX_URL = "https://webscraper.io/test-sites/e-commerce/ajax"  # target page
#AJAX_URL = "https://webscraper.io/test-sites"
AJAX_URL = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=1"
async def scrape_ajax_site():#scraping a
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)   # set False to watch it
        ctx = await browser.new_context()
        page = await ctx.new_page()
        await page.goto(AJAX_URL, timeout=60000)

        # Wait for initial product grid
        await page.wait_for_selector(".thumbnail", timeout=30000)

        # Keep clicking "Load more" until the button disappears or is disabled
        while True:
            try:
                # Button sometimes labeled "Load more" or has class .btn-load-more
                load_more = await page.query_selector("button:has-text('Load more')")
                if not load_more:
                    break
                is_disabled = await load_more.get_attribute("disabled")
                if is_disabled:
                    break

                # Click and wait for new items to appear
                before_count = len(await page.query_selector_all(".thumbnail"))
                await load_more.click()
                # Wait for network idle and more thumbnails to load
                await page.wait_for_load_state("networkidle")
                # Soft wait: poll until count increases or timeout
                for _ in range(30):
                    after_count = len(await page.query_selector_all(".thumbnail"))
                    if after_count > before_count:
                        break
                    await asyncio.sleep(0.2)
                # If no increase, assume end
                if after_count <= before_count:
                    break

            except Exception:
                # If the button isn't present or any error occurs, exit the loop
                break

        # Extract product cards
        cards = await page.query_selector_all(".thumbnail")
        rows = []

        for card in cards:
            # Title text and product link
            title_el = await card.query_selector(".title")
            title = (await title_el.text_content()).strip() if title_el else None
            url = await title_el.get_attribute("href") if title_el else None
            # Price
            price_el = await card.query_selector(".price")
            price = (await price_el.text_content()).strip() if price_el else None
            # Rating (count of .glyphicon-star)
            stars = await card.query_selector_all(".ratings .glyphicon-star")
            rating = len(stars) if stars else 0
            # Image
            img_el = await card.query_selector("img")
            img_src = await img_el.get_attribute("src") if img_el else None

            rows.append({
                "title": title,
                "price": price,
                "rating_stars": rating,
                "product_url": url,
                "image_url": img_src
            })

        await browser.close()
        return rows

# Run and save
data = asyncio.get_event_loop().run_until_complete(scrape_ajax_site())
print(f"Collected {len(data)} products")

# Save to CSV/JSON
Path("output").mkdir(exist_ok=True)
csv_path = Path("output/products_ajax.csv")
json_path = Path("output/products_ajax.json")

with open(csv_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)

with open(json_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"Saved CSV → {csv_path}")
print(f"Saved JSON → {json_path}")
````

````python
# Install necessary packages

!pip install playwright nest_asyncio # Install playwright and nest_asyncio libraries
!playwright install chromium # Install the Chromium browser driver for Playwright
!apt-get install libatk1.0-0 libatk-bridge2.0-0 libatspi2.0-0 libxcomposite1 # Install system dependencies required by Playwright on Linux

# Common imports and Colab event loop fix
import asyncio, json, csv, time # Import necessary modules: asyncio for async operations, json for JSON handling, csv for CSV handling, time for delays
from pathlib import Path # Import Path from pathlib for path manipulations
import nest_asyncio # Import nest_asyncio to allow nested use of asyncio event loops
nest_asyncio.apply() # Apply nest_asyncio to patch asyncio to allow reentrant loop calls, common in environments like Jupyter/Colab
from playwright.async_api import async_playwright # Import async_playwright for asynchronous browser automation

# Define the target URL for scraping
# AJAX_URL = "https://webscraper.io/test-sites/e-commerce/ajax"  # Original target page (commented out)
# AJAX_URL = "https://webscraper.io/test-sites" # Another example target (commented out)
AJAX_URL = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=1" # The current target URL for scraping laptop products

# Define an asynchronous function to scrape the AJAX-loaded site
async def scrape_ajax_site():
    # Use async_playwright as a context manager to ensure proper resource management
    async with async_playwright() as p:
        # Launch a Chromium browser instance in headless mode (set headless=False to see the browser UI)
        browser = await p.chromium.launch(headless=True)
        # Create a new browser context (useful for isolating browsing sessions)
        ctx = await browser.new_context()
        # Open a new page within the browser context
        page = await ctx.new_page()
        # Navigate to the target URL and wait up to 60 seconds for the page to load
        await page.goto(AJAX_URL, timeout=60000)

        # Wait for the initial product grid items (elements with class ".thumbnail") to be present on the page
        await page.wait_for_selector(".thumbnail", timeout=30000)

        # Loop indefinitely to click 'Load more' buttons until no more products can be loaded
        while True:
            try:
                # Attempt to find a 'Load more' button. It can be identified by its text or class.
                load_more = await page.query_selector("button:has-text('Load more')")
                # If no 'Load more' button is found, break the loop
                if not load_more:
                    break
                # Check if the 'Load more' button is disabled
                is_disabled = await load_more.get_attribute("disabled")
                # If the button is disabled, break the loop as there are no more items to load
                if is_disabled:
                    break

                # Get the count of products currently displayed before clicking 'Load more'
                before_count = len(await page.query_selector_all(".thumbnail"))
                # Click the 'Load more' button
                await load_more.click()
                # Wait for the network to be idle after clicking, indicating new content has potentially loaded
                await page.wait_for_load_state("networkidle")

                # Implement a soft wait: poll for a short period to see if the product count increases
                for _ in range(30):
                    # Get the new count of products after the click and wait
                    after_count = len(await page.query_selector_all(".thumbnail"))
                    # If new products have appeared, break this inner loop
                    if after_count > before_count:
                        break
                    # Wait for a short duration before re-checking
                    await asyncio.sleep(0.2)
                # If the product count did not increase (or stayed the same), assume no new products loaded and break the main loop
                if after_count <= before_count:
                    break

            except Exception: # Catch any exceptions that might occur (e.g., button not found after a certain point)
                # If an error occurs, it usually means the 'Load more' button is no longer available, so exit the loop
                break

        # Once all products are loaded, extract the details from each product card
        cards = await page.query_selector_all(".thumbnail") # Select all product cards
        rows = [] # Initialize a list to store extracted product data

        # Iterate through each product card found
        for card in cards:
            # Extract product title and its URL
            title_el = await card.query_selector(".title") # Find the title element
            title = (await title_el.text_content()).strip() if title_el else None # Get title text, strip whitespace, handle None if element not found
            url = await title_el.get_attribute("href") if title_el else None # Get the product URL from the href attribute

            # Extract product price
            price_el = await card.query_selector(".price") # Find the price element
            price = (await price_el.text_content()).strip() if price_el else None # Get price text, strip whitespace

            # Extract product rating (number of star icons)
            stars = await card.query_selector_all(".ratings .glyphicon-star") # Find all star icons within the ratings container
            rating = len(stars) if stars else 0 # Count the number of stars, default to 0 if none found

            # Extract product image URL
            img_el = await card.query_selector("img") # Find the image element
            img_src = await img_el.get_attribute("src") if img_el else None # Get the image source URL

            # Append the extracted details as a dictionary to the rows list
            rows.append({
                "title": title,
                "price": price,
                "rating_stars": rating,
                "product_url": url,
                "image_url": img_src
            })

        await browser.close() # Close the browser instance
        return rows # Return the list of scraped product data

# Run the asynchronous scraping function using the asyncio event loop
# `run_until_complete` blocks until the coroutine is finished
data = asyncio.get_event_loop().run_until_complete(scrape_ajax_site())
print(f"Collected {len(data)} products") # Print the total number of products collected

# Prepare to save the collected data to files
Path("output").mkdir(exist_ok=True) # Create an 'output' directory if it doesn't already exist
csv_path = Path("output/products_ajax.csv") # Define the path for the CSV output file
json_path = Path("output/products_ajax.json") # Define the path for the JSON output file

# Save data to a CSV file
with open(csv_path, "w", newline="", encoding="utf-8") as f: # Open the CSV file in write mode, ensuring no extra blank rows
    writer = csv.DictWriter(f, fieldnames=data[0].keys()) # Create a DictWriter, using keys from the first product as headers
    writer.writeheader() # Write the header row to the CSV
    writer.writerows(data) # Write all collected product data rows to the CSV

# Save data to a JSON file
with open(json_path, "w", encoding="utf-8") as f: # Open the JSON file in write mode
    json.dump(data, f, ensure_ascii=False, indent=2) # Dump the data to JSON, ensuring non-ASCII characters are preserved and output is human-readable with 2-space indentation

print(f"Saved CSV → {csv_path}") # Confirm CSV file saving and its path
print(f"Saved JSON → {json_path}") # Confirm JSON file saving and its path
````

## Create Scrapy Spider for Books

### Subtask:
Create a new Scrapy spider named 'books.py' in the 'ecommerce_scraper/spiders' directory. This spider will define the allowed domains and start URLs for 'books.toscrape.com'. It will also implement a `parse` method to extract the book title, price, rating, and product URL from each book, and include logic to follow pagination links to scrape data from all available pages.


**Reasoning**:
The first instruction is to create the necessary directory structure for the Scrapy spider.



````python
import os

# Define the directory path
spider_dir = "ecommerce_scraper/spiders"

# Create the directory structure if it doesn't exist
os.makedirs(spider_dir, exist_ok=True)
print(f"Directory '{spider_dir}' created successfully.")
````

**Reasoning**:
Now that the directory structure is in place, the next step is to create the `books.py` file within the `ecommerce_scraper/spiders` directory and populate it with the Scrapy spider code as detailed in the subtask instructions.



````python
import os

# Define the path for the spider file
spider_file_path = "ecommerce_scraper/spiders/books.py"

# Scrapy spider code content
spider_code = """import scrapy

class BooksSpider(scrapy.Spider):
    name = 'books'
    allowed_domains = ['books.toscrape.com']
    start_urls = ['http://books.toscrape.com/']

    def parse(self, response):
        # Helper dictionary to convert star rating text to numerical values
        rating_map = {
            'One': 1,
            'Two': 2,
            'Three': 3,
            'Four': 4,
            'Five': 5
        }

        # Iterate through each book article on the page
        for book in response.css('article.product_pod'):
            title = book.css('h3 a::attr(title)').get()
            price = book.css('.product_price .price_color::text').get()
            product_url = book.css('h3 a::attr(href)').get()

            # Extract rating text (e.g., 'star-rating Two')
            rating_text = book.css('.star-rating::attr(class)').get()
            rating_numeric = 0
            if rating_text:
                # Split the class string and get the rating word
                rating_word = rating_text.split()[-1]
                rating_numeric = rating_map.get(rating_word, 0)

            yield {
                'title': title,
                'price': price,
                'rating_stars': rating_numeric,
                'product_url': response.urljoin(product_url) # Ensure absolute URL
            }

        # Follow pagination links
        next_page = response.css('li.next a::attr(href)').get()
        if next_page is not None:
            yield response.follow(next_page, callback=self.parse)
"""

# Write the spider code to the file
with open(spider_file_path, "w") as f:
    f.write(spider_code)

print(f"Scrapy spider '{spider_file_path}' created successfully.")
````

````python
import os
import sys
import subprocess
import tempfile
import site

# 1. Ensure the 'output' directory exists
output_dir = "output"
os.makedirs(output_dir, exist_ok=True)

# 2. Get the absolute path of the current working directory
original_cwd = os.getcwd()

# 3. Define project_root and ensure the 'ecommerce_scraper' directory exists
project_root = "ecommerce_scraper"
os.makedirs(project_root, exist_ok=True)

# Define the path for the temporary script inside the project_root
temp_script_path = os.path.join(original_cwd, project_root, "run_spider_temp.py")

# Get the site-packages directory where scrapy should be installed
site_packages_path = site.getsitepackages()[0]
print(f"Scrapy is expected to be in: {site_packages_path}")

# 4. Construct the content for a temporary Python script
temp_script_content = f"""import os
import sys
from scrapy.cmdline import execute

# Add the original_cwd to sys.path within the script
if '{original_cwd}' not in sys.path:
    sys.path.insert(0, '{original_cwd}')

# Add site-packages path to sys.path within the script for scrapy module
if '{site_packages_path}' not in sys.path:
    sys.path.insert(0, '{site_packages_path}')

# Set the environment variable SCRAPY_SETTINGS_MODULE
os.environ['SCRAPY_SETTINGS_MODULE'] = 'ecommerce_scraper.settings'

# Change the working directory to the ecommerce_scraper project root
# This needs to be relative to where this temp script is located initially
os.chdir('{original_cwd}/{project_root}')

print('Executing Scrapy crawl from temporary script...')
# Call execute to run the spider and save its output
try:
    execute(['scrapy', 'crawl', 'books', '-o', '../output/books.csv'])
    print('Scrapy crawl completed successfully.')
except Exception as e:
    print(f'Error during Scrapy crawl: {{e}}')
"""

# 5. Create a temporary file named run_spider_temp.py and write the content
try:
    with open(temp_script_path, "w") as f:
        f.write(temp_script_content)
    print(f"Temporary Scrapy runner script created at: {temp_script_path}")

    # 6. Execute this run_spider_temp.py script using subprocess.run()
    print("Running temporary Scrapy script via subprocess...")
    process = subprocess.run(
        [sys.executable, temp_script_path],
        capture_output=True,
        text=True, # Capture output as text
        check=False # Do not raise an exception for non-zero exit codes immediately
    )

    # 7. Print the standard output and error from the subprocess execution
    print("\n--- Subprocess Stdout ---")
    print(process.stdout)
    print("\n--- Subprocess Stderr ---")
    print(process.stderr)

    if process.returncode != 0:
        print(f"Subprocess finished with non-zero exit code: {process.returncode}")
        raise subprocess.CalledProcessError(process.returncode, process.args, output=process.stdout, stderr=process.stderr)

except subprocess.CalledProcessError as e:
    print(f"Error executing temporary script: {e}")
    print(f"Stdout: {e.stdout}")
    print(f"Stderr: {e.stderr}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
finally:
    # 8. Ensure the temporary script file is removed
    if os.path.exists(temp_script_path):
        os.remove(temp_script_path)
        print(f"Removed temporary script: {temp_script_path}")

    # 9. Change the working directory back to the original_cwd
    os.chdir(original_cwd)
    print(f"Changed back to original directory: {os.getcwd()}")


````
